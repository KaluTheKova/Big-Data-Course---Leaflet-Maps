```python
## IMPORTS
from IPython.display import display as disp
from ipywidgets import Layout, HTML
from ipyleaflet import CircleMarker, Map
import urllib
import json
import time

### PART 1
#1. Read the current bus locations in Tampere
#2. Add one marker per bus to the map, using the bus coordinates.
#3. Sleep 5 sec on every run.

## Starting stuff and url.
busDataUrl = "http://data.itsfactory.fi/journeys/api/1/vehicle-activity"
dataFetchFlag = True
dataCap = int(input("Set how many times program will run. Leave blank for default (3).") or "3")


# Center coordinates
center = (61.491111, 23.794797)
m = Map(center=center, zoom=13, layout=Layout(width='900px', height='600px'))
disp(m)

# Main loop & mainloop counter.
dataCount = 0
while dataCap > dataCount:
    dataCount += 1
    data = json.loads(urllib.request.urlopen(busDataUrl).read())
    
    # Fetch busdata
    for obj in data["body"]:
        o = obj["monitoredVehicleJourney"]
        vehicleRef = o["vehicleRef"]
        latitude = o["vehicleLocation"]["latitude"]
        longitude = o["vehicleLocation"]["longitude"]
        
        # Set marker location and popup info
        marker = CircleMarker(location=(latitude, longitude), radius=5,
                      popup=HTML("Busline is: " + str(vehicleRef)))
        
        # Add a marker layer
        m.add_layer(marker);
        
    # Wait 5 secs.
    time.sleep(5)
```

```python
### PART 3

#1. Embed geographic boundaries of all municipalities in Finland to the map.
   # Use the file municipalities.geojson in similar manner
   # as the example used regions.geojson.
   
#2. Read the data about average housing areas.

#3. Color the municipalities according to the housing area values.
   # If you inspect municipalities.geojson, you may note that the "key_on" parameter
   # for choropleth should be "feature.properties.name".

from IPython.display import display as disp
import pandas as pd
import folium
import json

with open("regions.geojson") as infile:
    boundjson = json.load(infile)

m = folium.Map(location=[65.3, 25.5], zoom_start=5, width=800, height=600)
bounds = folium.GeoJson(boundjson)
bounds.add_to(m)

testdata = pd.DataFrame([["Pirkanmaa", 100], ["Lappi", 200]])

m.choropleth(
    geo_data=boundjson,
    name='choropleth',
    data=testdata,
    columns=[0, 1],
    key_on='feature.properties.Maakunta',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    highlight=True)


folium.LayerControl().add_to(m)
disp(m)
```
